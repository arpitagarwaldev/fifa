```python
import sqlite3
import pandas as pd
```

```python
# Load the CSV file into a pandas DataFrame
csv_file = '/Users/arpitagarwal/Desktop/coding/end-2-end ml project/notebook/data/player_stats.csv'
df = pd.read_csv(csv_file, encoding='ISO-8859-1')

```

```python
# Connect to an SQLite database (it will create the database if it doesn't exist)
conn = sqlite3.connect('/Users/arpitagarwal/Desktop/coding/end-2-end ml project/notebook/data/player_stats.db')

# Convert the DataFrame to SQLite
df.to_sql('player_stats', conn, if_exists='replace', index=False)

# Commit and close the connection
conn.commit()
conn.close()

print("CSV successfully converted to SQLite3 database!")
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('data/player_stats.db')

# Create a cursor object to interact with the database
cursor = conn.cursor()

# Query to fetch all data from the 'player_stats' table
cursor.execute("SELECT * FROM player_stats")

# Fetch all rows of the query result
rows = cursor.fetchall()

# Print the data
for row in rows:
    print(row)

# Close the connection
conn.close()
```

```python
# Connect to the SQLite database
conn = sqlite3.connect('data/player_stats.db')

# Create a cursor object
cursor = conn.cursor()

# Query to fetch the table's data
cursor.execute("SELECT * FROM player_stats")

# Get the column headers from the cursor description
headers = [description[0] for description in cursor.description]

# Print the headers
print(headers)

# Close the connection
conn.close()
```
